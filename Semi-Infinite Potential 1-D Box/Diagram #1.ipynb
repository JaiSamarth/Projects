```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,ax = plt.subplots(figsize=(6,5),dpi=300)
plt.style.use('seaborn-paper')
ax.fill_between([-1,0],[5,5],facecolor='b',alpha=0.25)
ax.fill_between([2,5],[2,2],facecolor='b',alpha=0.25)
ax.plot([0,0],[0,5],'k--')
ax.plot([2,2],[0,2],'k--')
ax.plot([2,5],[2,2],'k--')
ax.plot([-1,-1],[0,5],'k-')
ax.plot([5,5],[0,5],'k-')
ax.plot([-1,5],[5,5],'k-')
ax.plot([-1,5],[0,0],'k-')
ax.set_ylim(-0.01,5.01)
ax.set_xlim(-1.01,5.01)
ax.set_axis_off()
plt.annotate(r'$V_o$',xy=(2,2),xytext=(2,1),size=20,arrowprops=dict(facecolor='k',shrink=0))
plt.annotate(r'$V_o$',xy=(2,0),xytext=(2,1),size=20,arrowprops=dict(facecolor='k',shrink=0))
plt.annotate(r'$\infty$',xy=(0,5),xytext=(0,3),size=20,arrowprops=dict(facecolor='k',shrink=0))
plt.annotate(r'$\infty$',xy=(0,0),xytext=(0,3),size=20,arrowprops=dict(facecolor='k',shrink=0))
plt.annotate(r'$a$',xy=(2,2),xytext=(1,2.09),size=20,arrowprops=dict(facecolor='k',shrink=0))
plt.annotate(r'$a$',xy=(0,2),xytext=(1,2.09),size=20,arrowprops=dict(facecolor='k',shrink=0))
plt.savefig('Box.png')
```
